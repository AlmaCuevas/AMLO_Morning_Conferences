```python
import os
import pandas as pd
import seaborn as sns
```

```python
root_dir = os.getcwd()
year_folders = [f for f in os.listdir(root_dir) if f.startswith("20")]

all_df = pd.DataFrame(
    columns=["Participante", "Texto", "Sentimiento", "Palabras", "Dia", "Mes", "Anio"]
)
```

```python
# Tree of folders
for year in year_folders:
    month_folders = os.listdir(os.path.join(root_dir, year))
    for month in month_folders:
        day_folders = os.listdir(os.path.join(*[root_dir, year, month]))
        for day in day_folders:
            final_path = os.path.join(*[root_dir, year, month, day])
            file_name = [f for f in os.listdir(final_path) if f.endswith(".csv")][0]
            all_df = pd.concat(
                [
                    pd.read_csv(os.path.join(*[root_dir, year, month, day, file_name])),
                    all_df,
                ],
                axis=0,
                ignore_index=True,
            )
all_df
```

# People interventions count

```python
import re
def limpiar_texto(texto):
    """
    Función para realizar una limpieza de un texto dado.
    """
    # Eliminamos los caracteres especiales
    texto = re.sub(r'\W', ' ', str(texto))
    # Eliminado las palabras que tengo un solo caracter
    texto = re.sub(r'\s+[a-zA-Z]\s+', ' ', texto)
    # Sustituir los espacios en blanco en uno solo
    texto = re.sub(r'\s+', ' ', texto, flags=re.I)
    # Convertimos textos a minusculas
    texto = texto.lower()
    if "lopez obrador" in texto:
        texto = "President"
    elif "pregunta" in texto:
        texto = "Public"
    else:
        texto = "Councils"
    return texto
```

```python
all_df['Year'] = all_df['Anio']
all_df['Participant'] = all_df['Participante'].apply(limpiar_texto)
all_df['Participant']
```

```python
words_per_year = all_df.groupby(by=["Participant", "Year" ])["Participant"].count()
words_per_year = words_per_year.unstack(level=1)
words_per_year
```

```python
ax = words_per_year.plot(kind = 'bar')
ax.set_ylabel('Number of participations')
```

```python
ax = sns.boxplot(data=words_per_year.unstack(level=1), x="Participante_limpio", hue="Year", palette="husl", dodge=True)
ax.set(xlabel='Participants',
       ylabel='Number of participations',
       title='Participations per year')
```
